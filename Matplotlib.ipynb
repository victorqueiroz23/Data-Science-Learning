## <center> <b>Estrutura do Matplotlib</b> </center>

##### <b>Figura:</b> Apresentação de um resultado. Pode conter vários painéis (útil para DASHBOARDS!). 
##### <b>Axes:</b> É o painel da figura! Cada um tem um título, seus eixos, rótulos e intervalos dos eixos.
##### <b>Axis:</b> São os eixos (x, y) de cada painél. Cada eixo tem seu rótulo (LABELS: <b>x_label</b>, <b>y_label</b>)

![](https://matplotlib.org/1.5.1/_images/fig_map.png)

![](https://matplotlib.org/stable/_images/sphx_glr_anatomy_001_2_0x.png)


```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig_example, fig_example_axes = plt.subplots(ncols = 2, nrows = 2, figsize=(8, 6))
fig_example_axes
```

```python
# Estrutura da Figura:           ncol = 1     ncol = 2     ...     ncol = N
#                              _________________________         ____________
#                    nrow = 0 | Painel_0_0 | Painel_0_1 |  ...  | Painel_0_N |  
#                             |____________|____________|       |____________|
#                    nrow = 1 | Painel_1_0 | Painel_1_1 |  ...  | Painel_1_N |
#                             |____________|____________|       |____________|                
#                                  ...         ...
#                              _________________________         ____________
#                    nrow = M | Painel_M_0 | Painel_M_1 |  ...  | Painel_M_N |
#                             |____________|____________|       |____________|

#-----------------------------------------------------------------------------------------------------------------

x = np.arange(0, 10, 0.1)

# Criando figura e definindo o número de painéis (matrix contendo nº de linhas = nrwos, e nº de colunas = ncols)
fig_1, fig_1_axes = plt.subplots(ncols = 2, nrows = 2, figsize = (15, 10))
fig_1.suptitle("Gráficos de Relações Trigonométricas", size = 30)  # Título da Figura (com tamanho de fonte = 30!

# Painel 1
fig_1_axes[0, 0].plot(np.sin(x), label = 'sen(x)') # Primeiro desfecho
fig_1_axes[0, 0].plot(np.cos(x), label = 'cos(x)') # Segundo desfecho
fig_1_axes[0, 0].set_title("Gráfico de Seno e Cosseno") # Título do Painel 1
fig_1_axes[0, 0].legend()   # Nomes dos labels definidos agora aparecerão no Painel 1!

# Painel 2
fig_1_axes[0, 1].plot(np.cos(x), label = 'cos(x)', color = 'purple') # Primeiro desfecho
fig_1_axes[0, 1].plot(np.sin(x), label = 'sen(x)', color = 'red') # Segundo desfecho
fig_1_axes[0, 1].set_title("Gráfico de Cosseno e Seno") # Título do Painel 2
fig_1_axes[0, 1].legend()

# Painel 3
fig_1_axes[1, 0].plot(np.tan(x))
fig_1_axes[1, 0].set_title("Gráfico de Tangente")

# Painel 4
fig_1_axes[1, 1].plot(np.cosh(x))
fig_1_axes[1, 1].set_title("Gráfico de Cossencante")
```

#### Método GROUPBY: df<b>.groupby(</b>categoria_maior, categoria_menor_1, ... , categoria_menor_N<b>)</b><i>.operação()</i>
##### <i>OBS:</i> Categoria = Parâmetro (coluna)!

```python
# MÉDIA DAS NOTAS DE ACORDO COM RENDA FAMILIAR, EM CADA ESTADO -> DATAFRAME
#-------------------------------------------------------------------------
df_ENEM_Q006.filter(items = ['SG_UF_ESC', 'NU_NOTA_CN', 'Q6_RESP'])\
       .groupby(['SG_UF_ESC', 'Q6_RESP']).mean()
```

#### <i>OBS:</i> Manipulando a estrutura dos painéis -> <b>https://matplotlib.org/3.1.0/api/axes_api.html?highlight=axes#module-matplotlib.axes</b>

```python
fig, ax = plt.subplots(figsize=(12, 10))

df_ENEM_Q006.filter(items = ['SG_UF_ESC', 'NU_NOTA_REDACAO', 'Q006'])\
       .groupby(['Q006', 'SG_UF_ESC']).mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)\
       .unstack().plot(ax=ax)

       
ax.set_title('Nota Redação x Renda Familiar x Estado', fontsize = 20) # Definimos o nome da figura de painel único!
ax.legend(bbox_to_anchor = (1.05, 1), loc = 0, borderaxespad = 0.)    # Definimos onde a caixa de legenda vai ficar!
```
