<h3><b>Importando bibliotecas</b></h3>


```python
import requests as req
from bs4 import BeautifulSoup as BS
import pandas as pd
```

<h3><b> Requisição para o site </b></h3>

<p>1º) <b>Site:</b> https://finance.yahoo.com/cryptocurrencies/?offset=0&count=100' </p>

<p> 2º) Definir quantas criptomoedas serão visualizadas em <b>count=</b>[...] </p>


```python
# Diretório fonte da cotação das criptomoedas:
url = 'https://finance.yahoo.com/cryptocurrencies/?offset=0&count=100'

# Fazendo a requisição com requests!
pagina_html = req.get(url).text

# Status da requisição:
pagina_html[:100]
```

<h3><b> Transformando em objeto do BeautifulSoup </b></h3>

```python
#Lendo o HTML com BS
yahoo_cripto_bs = BS(pagina_html, 'html.parser')
```

<h3><b> Gerando dataframe vazio só com o cabeçalho </b></h3>
<p> dataframe: <b>df_crypto</b>

```python
df_crypto = pd.DataFrame(columns = ['name', 'price_intraday', 'change', '%_change', 'market_cap'])
```

<h3><b> Localizando a tabela no objeto BS </b></h3>
<p> 1º) Tag: <b>tbody</b> <p>
<p> 2º) Passando dados para <b> df_crypto </b>



```python
tabela = yahoo_cripto_bs.find('tbody').find_all('tr')

#contador = 0

for linha in tabela:
  colunas = linha.find_all('td')
  
  #Definindo onde estão dados por linha
  name = colunas[1].text
  price = colunas[2].text
  change = colunas[3].text
  change_per = colunas[4].text
  market_cap = colunas[5].text
  
  #print('Linha', contador, '->', name, '|', price, '|', change, '|', change_per, '|', market_cap, '|')
  #contador += 1

  df_crypto = df_crypto.append({'name': name, 'price_intraday': price, 'change': change, '%_change': change_per, 'market_cap': market_cap}, ignore_index=True)
```
